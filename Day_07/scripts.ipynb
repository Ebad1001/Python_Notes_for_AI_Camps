## Working with Text Files

- Reading from a text file

````python
file = open("./data/story.txt", "r")

text = file.read()
print(text)

file.close()
````

- Text formatting

````python
# removing unwanted characters from the text
unwanted_char = [',', '.', ';', ':', '!', '?', '\'', '\"', '\t', '\n', '  ']
for char in unwanted_char:
    text = text.replace(char, ' ')
text = text.strip()
print(f"Clean text = {text}|")

# converting text to lowercase
text = text.lower()
print(f"lowered = {text}")

# finding frequencies for each letter
letter_frequency = {}
for letter in text:
    letter_frequency[letter] = text.count(letter)
print(f"Letter frequencies = {letter_frequency}")

# splitting text into words
words = text.split(' ')
print(f"Words = {words}")

# finding unique words
unique = set(words)
print(f"Unique words = {unique}")

# finding word count
print(f"Total word count = {len(words)}")
print(f"Unique word count = {len(unique)}")

# finding frequencies for each word
word_frequency = {}
for word in unique:
    word_frequency[word] = words.count(word)
print(f"Word frequencies = {word_frequency}")
````

- Writing into a text file

````python
file = open("./data/output.txt", "w")

file.write("Word         : Frequency\n")
for word in word_frequency:
    file.write(f"{word:12} : {word_frequency[word]}\n")

file.close()
````

````python
file = open("./data/output.txt", "w")

file.write("Letter : Frequency\n")
for letter in letter_frequency:
    file.write(f"{letter} : {letter_frequency[letter]}\n")

file.close()
````

## Working with Pandas DataFrames

````python
import pandas as pd

table = pd.read_csv("./data/dataset.csv")
table.head()
````

To access particular rows, columns or cells - we use the following syntax:

```Python
    # Syntax:
    table.loc[row_filter, col_filter]
```

If you don't want to apply a filter, put `:` instead

```Python
    # Syntax:
    table.loc[:, col_filter]
    table.loc[row_filter, :]
```

- accessing specific a row (using row index)

````python
my_row = table.loc[0, :]

print(my_row)
````

- accessing specific a column (using column name)

````python
my_col = table.loc[:, 'sepal length']

print(my_col)
````

- accessing specific cells

````python
my_cell = table.loc[2, 'sepal length']

print(my_cell)
````

- filtering rows based on condition

````python
filtered = table[table['sepal length'] > 7, :]

print(filtered)
````

- sorting rows based on columns

````python
# single column
sorted = table.sort_values(by='sepal length', ascending=True)

print(sorted)
````

````python
# multiple columns
sorted = table.sort_values(by=['sepal length', 'sepal width'], ascending=[True, False])

print(sorted)
````
