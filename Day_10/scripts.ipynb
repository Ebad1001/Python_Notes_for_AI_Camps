**API** = `Application Programming Interface`

An API allows different software applications to talk to each other. It acts as a middleman, taking a request from one program, delivering it to another, and bringing the response back



- Lets some examples to understand the use of APIs

```python
import requests
import json

country_name = "india"
response = requests.get(f"https://restcountries.com/v3.1/name/{country_name}")
json_data = json.loads(response.text)

with open("output.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)
```

```python
import requests
import json

latitude, longitude = 28.61, 77.23
response = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true")
json_data = json.loads(response.text)

with open("output.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)
```

- Some more examples of APIs that we can use to make useful apps

```python
import requests
import json

amount = 1
convert_from = "INR"
convert_to = "USD"

url = f"https://api.exchangerate-api.com/v4/latest/{convert_from}"
response = requests.get(url)
json_data = json.loads(response.text)

with open("output.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)
    
multiplier = json_data["rates"][convert_to]
answer = amount * multiplier

print(f"{amount} {convert_from} is equal to {answer} {convert_to}")
```

```python
import requests
import json

word = "artificial intelligence"
response = requests.get(f"https://api.dictionaryapi.dev/api/v2/entries/en/{word}")
json_data = response.json()

with open("output.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)

print("Word:", json_data[0]["word"])
print("Meaning:", json_data[0]["meanings"][0]["definitions"][0]["definition"])
```

```python
import requests
import json

username = "Ebad1001"
url = f"https://api.github.com/users/{username}"

response = requests.get(url)
json_data = response.json()

with open("output.json", "w", encoding="utf-8") as f:
    json.dump(json_data, f, ensure_ascii=False, indent=4)

if response.status_code == 200:
    data = response.json()
    
    print("Username:", data["login"])
    print("Name:", data["name"])
    print("Public Repos:", data["public_repos"])
    print("Followers:", data["followers"])
    print("Following:", data["following"])
else:
    print("Error:", response.status_code)
```
