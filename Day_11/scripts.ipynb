- To interact with Google's various LLMs, you can use an Google API Key.
- To get a free API key, visit: [aistudio.google.com](https://aistudio.google.com/app/apikey)

````python
import google.generativeai as genai

# replace this with your API key
API_KEY="AIza..."
````

````python
# configure genai
genai.configure(api_key=API_KEY)

# set up the model, "gemini-2.5-flash" is a good model to start with
model = genai.GenerativeModel(model_name="gemini-2.5-flash")

# start a conversation
chat = model.start_chat(history=[])
````

````python
print("Welcome to the Gemini Chatbot!\n")

request = input("You: ")
response = chat.send_message(request).text

print(f"User:    {request}")
print(f"Chatbot: {response}")
````

- Now writing your API key directly into the code is not a good idea.
- You should use environment variables to store your API key.
- To do this, create a file named `.env` in your project folder.

```python
    # .env file
    API_KEY="AIza..."
```

- Then, in your Python code, import the `load_dotenv` function from the `dotenv` module and call the `load_dotenv` function, as follows:

````python
import os
from dotenv import load_dotenv

load_dotenv(override=True) 
API_KEY = os.getenv("API_KEY")
print(API_KEY)

# If you can see your API key in the output, you are good to go!
````
